```python
country ='all'
indicator = 'NY.GDP.PCAP.PP.KD'
URL = f"https://api.worldbank.org/v2/country/{country}/indicator/{indicator}"
params = {
    'format':'json'}

response = requests.get(URL, params = params)
total = response.json()[0]['total']
params = { 'format':'json',
          'per_page': total
}
response = requests.get(URL, params = params)
gdp_per_capita = pd.json_normalize(response.json()[1])
gdp_per_capita = gdp_per_capita.rename(columns = {'value':'gdp_per_capita'})
gdp_per_capita.head(2)
```

**Q2. Now, use the API to get all available data for Life expectancy at birth, total (years).** **This indicator has code "SP.DYN.LE00.IN".**  **Again, convert the results to a DataFrame.**

```python
country ='all'
indicator = 'SP.DYN.LE00.IN'
URL = f"https://api.worldbank.org/v2/country/{country}/indicator/{indicator}"
params = {
    'format':'json'}
response = requests.get(URL, params = params)
total = response.json()[0]['total']
params = { 'format':'json',
          'per_page': total
}
response = requests.get(URL, params = params)
life_expectancy = pd.json_normalize(response.json()[1])
life_expectancy = life_expectancy.rename(columns = {'value':'life_expectancy'})
life_expectancy.head(2)
```

**Q3. Merge the two results DataFrames together. You may want to rename or drop columns prior to merging.**

```python
gdp_df = gdp_per_capita
```

```python
gdp_df
```

```python
le_df = life_expectancy
```

```python
le_df
```

```python
gdp_df = (
         gdp_df
         .drop(columns = ['unit', 'obs_status', 'decimal', 'indicator.id', 'indicator.value'])
)
```

```python
gdp_df
```

```python
le_df = (
         le_df
         .drop(columns = ['unit', 'obs_status', 'decimal', 'indicator.id', 'indicator.value'])
)
```

```python
le_df
```

```python
gdp_le_df = pd.merge(
                     left = gdp_df,
                     right = le_df,
                     how = 'inner',
                     validate ='one_to_one'
)
```

```python
gdp_le_df.head(2)
```

**Q4. You can also get more information about the available countries (region, capital city, income level classification, etc.) by using the Country API.** **Use this API to pull in all available data. Merge this with your other datasets. Use this to now remove the rows that correspond to regions and not countries.**

```python
base_URL = 'https://api.worldbank.org/v2/country'
params = {'format': 'json'}
response = requests.get(base_URL, params=params) 
data = response.json()
country_df = pd.json_normalize(data[1]) 
country_df.head(2)
```

```python
gdp_le_country_df = pd.merge(gdp_le_df, country_df, left_on='countryiso3code', right_on='id', how='left', validate = 'many_to_one'
)
filtered_df = gdp_le_country_df[gdp_le_country_df['region.id'] != 'REG']
filtered_df.head(2)
```
